```python
import graphviz

# Inicialização do grafo
der = graphviz.Digraph(format='png')
der.attr(rankdir='LR', size='10')

# Entidades e atributos
entities = {
    "Usuário": [
        "id_usuario (PK)", "nome", "email", "senha_hash",
        "cargo", "nivel_acesso", "ativo"
    ],
    "Área": [
        "id_area (PK)", "nome_area", "descricao"
    ],
    "Veículo": [
        "id_veiculo (PK)", "nome_modelo", "descricao"
    ],
    "Risco": [
        "id_risco (PK)", "descricao", "data_registro",
        "nivel_criticidade", "status",
        "id_usuario (FK)", "id_area (FK)", "id_veiculo (FK)"
    ],
    "Solução": [
        "id_solucao (PK)", "descricao", "efetividade_estimativa", "id_risco (FK)"
    ],
    "Histórico de Riscos": [
        "id_historico (PK)", "id_risco (FK)", "data_alteracao",
        "descricao_antes", "descricao_depois", "alterado_por (FK)"
    ],
    "Sugestão IA": [
        "id_sugestao (PK)", "tipo", "conteudo", "id_risco_relacionado (FK)"
    ],
    "Alerta": [
        "id_alerta (PK)", "id_risco (FK)", "data_envio",
        "status_entrega", "destinatario (FK)"
    ]
}

# Adicionando entidades ao grafo com estilo de tabela
for entity, attributes in entities.items():
    label = f"<<TABLE BORDER='1' CELLBORDER='1' CELLSPACING='0'>"
    label += f"<TR><TD BGCOLOR='lightblue'><B>{entity}</B></TD></TR>"
    for attr in attributes:
        label += f"<TR><TD>{attr}</TD></TR>"
    label += "</TABLE>>"
    der.node(entity, label=label, shape='plaintext')

# Relacionamentos
der.edge("Risco", "Usuário", label="id_usuario (FK)")
der.edge("Risco", "Área", label="id_area (FK)")
der.edge("Risco", "Veículo", label="id_veiculo (FK)")
der.edge("Solução", "Risco", label="id_risco (FK)")
der.edge("Histórico de Riscos", "Risco", label="id_risco (FK)")
der.edge("Histórico de Riscos", "Usuário", label="alterado_por (FK)")
der.edge("Sugestão IA", "Risco", label="id_risco_relacionado (FK)")
der.edge("Alerta", "Risco", label="id_risco (FK)")
der.edge("Alerta", "Usuário", label="destinatario (FK)")

# Geração do arquivo
# Renderiza o diagrama para PNG
der.render('diagrama_der', format='png')

# Exibe a imagem no próprio notebook
from IPython.display import Image
Image(filename='diagrama_der.png')
```
